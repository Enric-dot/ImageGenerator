```python
def build_spiral(t, length=50, angle=4, reduction=2, min_size=10, counter=-1, stop=200):
    if counter != -1:
        counter += 1
    if (length > min_size) and (counter != stop):
        t.forward(length)
        t.left(angle)
        build_spiral(t, float(length / reduction), angle, reduction, min_size, counter, stop)
    else:
        t.penup()

```

```python
MINIMUM_BRANCH_LENGTH = 5

turtle.initializeTurtle(initial_speed=10)
turtle.bgcolor('black')
turtle.hideturtle()
turtle.penup()
starty =200 #@param {type:"number"}
startx = 400#@param {type:"number"}
width = 1#@param {type:"number"}
turtle.width(width)
turtle.sety(starty)
turtle.setx(startx)
turtle.pendown()
turtle.setheading(90)
color =  255,100,0#@param {type:"raw"}
turtle.color(color)
length =  200#@param {type:"number"}
reduction =  1.004#@param {type:"number"}
angle = 360/6+0.25 #@param {type:"raw"}
build_spiral(turtle, length=length, angle=angle, reduction=reduction, min_size=3, counter=0, stop=600)
```
